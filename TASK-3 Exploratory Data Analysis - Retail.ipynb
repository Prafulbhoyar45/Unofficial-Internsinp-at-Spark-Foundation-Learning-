# EDA (Explorative Data Analysis)

```python
# Pairplot
# To see is there any liniarity between columns
sns.pairplot(data_retail)
plt.show()
```

```python
# Lets conclude the correlation between the veriabls
# By using heatmap
data_retail.corr()
```

```python
# map that correlation
sns.heatmap(data_retail.corr(),annot=True)
plt.show()
```

```python
# Find how may catagoaries of supplies we haven in retail
data_retail.Category.value_counts()
```

```python
# Visualise the percentage
plt.figure(figsize=(6,6))
textprops = {'fontsize':15}
plt.title('Category')
plt.pie(data_retail['Category'].value_counts(), labels=data_retail['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

```python
# Find by which Categories we have most profit
data_retail.groupby('Category')['Profit','Sales'].agg('sum')
```

```python
# Visualise by which Categories we have most profit
data_retail.groupby('Category')['Profit','Sales'].agg('sum').plot(kind = 'bar')
plt.show()
```

```python
# Computing top categories in terms of sales and profit from first 100 observation and plotting them.
```

```python
top_category_s = data_retail.groupby('Category').Sales.sum().nlargest(n=100)
top_category_p = data_retail.groupby('Category').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
top_category_s.plot(kind='bar', figsize=(10,5), fontsize=14)
top_category_p.plot(kind='bar', figsize=(10,5), fontsize=14, color='red')
plt.xlabel('Category', fontsize=15)
plt.ylabel('Total Sales/Profits', fontsize=15)
plt.title('Top Category Sales vs Profit', fontsize=15)
plt.show()
```

```python
# Computing top sub-categories in terms of sales and profit from first 100 observation and plotting them.
```

```python
top_subcategory_s = data_retail.groupby('Sub-Category').Sales.sum().nlargest(n=100)
top_subcategory_p = data_retail.groupby('Sub-Category').Profit.sum().nlargest(n=100)
plt.style.use('seaborn')
top_subcategory_s.plot(kind='bar', figsize=(10,5), fontsize=14)
top_subcategory_p.plot(kind='bar', figsize=(10,5), fontsize=14, color='red')
plt.xlabel('Sub-Category', fontsize=15)
plt.ylabel('Total Sales/Profits', fontsize=15)
plt.title('Top Sub-Category Sales vs Profit', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(14,12))
statewise = data_retail.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh()
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.lineplot(data_retail['Discount'], data_retail['Profit'], data=data_retail)
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.title('Region')
plt.pie(data_retail['Region'].value_counts(), labels=data_retail['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
# The graph above shows that West and East have same profit though sales in the East are less as compared to West.

# Visualising the Sales vs Profits in different states
```

```python
top_states_s = data_retail.groupby("State").Sales.sum().nlargest(n=10)
top_states_p = data_retail.groupby("State").Profit.sum().nlargest(n = 10)
plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 states Sales vs Profit",fontsize = 15)
plt.show()
```

```python
plt.style.use('seaborn')
data_retail.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "viridis")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)
plt.show()
```

```python
# The graph clearly shows that if we giive more Discount on our products sales increases but profit decreases.\
```
